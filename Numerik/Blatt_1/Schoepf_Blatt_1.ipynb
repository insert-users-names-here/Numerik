**Aufgabe 2**

```python


def f(x_eins):
    return 5*(3**x_eins)

def g(x_zwei):
    return 2*(x_zwei**3)

x = np.linspace(0.1,10, 400)

y_eins = []
y_zwei = []

for i in x:
    y_eins.append(f(i))

for i in x:
    y_zwei.append(g(i))

plt.figure(figsize=(15, 5))


#normal geplottet

plt.subplot(1, 3, 1)
plt.plot(x, y_eins, label="f(x) = 5 * 3^x")
plt.plot(x, y_zwei, label="g(x) = 2 * x^3 ")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc="upper left")
plt.title("normal geplottet")


# einfach logarithmisch

plt.subplot(1, 3, 2)
plt.semilogy(x, y_eins, label="f(x) = 5 * 3^x")
plt.semilogy(x, y_zwei, label="g(x) = 2 * x^3 ")
plt.xlabel('x')
plt.ylabel('log(y)')
plt.legend(loc="upper left")
plt.title("einfach logarithmische Skalierung")


# doppelt logarithmisch

plt.subplot(1, 3, 3)
plt.loglog(x, y_eins, label="f(x) = 5 * 3^x")
plt.loglog(x, y_zwei, label="g(x) = 2 * x^3 ")
plt.legend(loc="upper left")
plt.xlabel('log(x)')
plt.ylabel('log(y)')
plt.title("doppelt logarithmische Skalierung")


plt.show()

```

**zu 2 (a) und (b):**

Es ist ersichtlich, dass durch die einfache Logarithmierung deine gerade aus Funktion f(x) entsteht da: 
$$
log(f(x))= log(5) + x\cdot log(3)
$$
bei g ist dies hingegen nicht der Fall. erst durch die logarithmische skalierung der x Achse entsteht eine Geradenfunktion bezglich log(x):
$$
log(g(x))= log(2)+ 3\cdot log(x)
$$

Dies gilt fr allgemein fr exponential und ptenzfunktionen. Die Begrndung liegt wie oben gezeigt darin, dass eine Exponentialfunktion mit einfach logarithmischer Skalierung linear ist, durch den Exponenten der durch die Logarithmus Rechenregeln nach unten kommt. Die Potenzfunktion ist erst linear wenn man gegen log(x) auftrgt.

```python
# (a)

import numpy as np
import matplotlib.pyplot as plt

def Folge_berechnen(x0, N):
    Folge = [x0]
    for i in range(1, N):
        xn_plus_eins = 0.5*(Folge[-1] + (2/Folge[-1]))
        Folge.append(xn_plus_eins)
    return Folge

def Zeichne(x0, N):
    Folgenglied = Folge_berechnen(x0, N)
    n = np.arange(N)

    plt.plot(n, Folgenglied)
    plt.show()

Zeichne(0.0001,90)
```

```python
# (b)

def Finde_n0(x0, epsilon):
    sqrt_2 = np.sqrt(2)
    xn = x0
    n = 0

    while abs(xn - sqrt_2) >= epsilon:
        xn = 0.5*(xn+(2/xn))
        n += 1 

    return n     

Finde_n0(0.0000001,0.001)
```
